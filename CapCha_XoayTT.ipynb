```python
def graft(img1,img2):
    img1[68:174+105,68:174+105]= img1[68:174+105,68:174+105] + img2
    a = img1
    return a
plt.imshow(graft(img1,img2))
```

```python
def rotate(img,x):
    (h, w) = img.shape[:2]
    (cX, cY) = (w // 2, h // 2)
    M = cv2.getRotationMatrix2D((cX, cY), x, 1.0)
    rotated = cv2.warpAffine(img, M, (w, h))
    return rotated
```

```python

for i in range(20,160):
    rotate(img1,i)

for i in range(20,160):
    rotate(img1,-i)
```

```python
arr_temp = np.zeros((347,347))
x = cv2.circle(arr_temp,(174,174),106,1,2)
angle , score = 0,999999999
for j in range(20,160):
    a = rotate(img1,j)
    b = rotate(img2,-j)
    g = graft(a,b)
    g = cv2.Canny(g,200,200)
    m = g*x
    m = m.sum()
    if (m < score):
        score = m
        angle = j
print(angle,score)
```
